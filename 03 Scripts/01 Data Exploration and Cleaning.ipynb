### Data Cleaning

```python
# Checking for duplicates

df.duplicated().any()
```

```python
# Renaming columns

df.rename(columns={'Life expectancy ': 'Life_expectancy'}, inplace=True)
df.rename(columns={'Adult Mortality': 'Adult_mortality'}, inplace=True)
df.rename(columns={'infant deaths': 'Infant_mortality'}, inplace=True)
df.rename(columns={'Alcohol': 'Alcohol_consumption'}, inplace=True)
df.rename(columns={'percentage expenditure': 'Health_expenditure'}, inplace=True)
df.rename(columns={'Hepatitis B': 'HepB_immunization'}, inplace=True)
df.rename(columns={'Measles ': 'Measles_cases'}, inplace=True)
df.rename(columns={'under-five deaths ': 'Under_5_deaths'}, inplace=True)
df.rename(columns={' BMI ': 'BMI'}, inplace=True)
df.rename(columns={'Polio': 'Polio_immunization'}, inplace=True)
df.rename(columns={'Total expenditure': 'Total_health_expenditure'}, inplace=True)
df.rename(columns={'Diphtheria ': 'Diphtheria_immunization'}, inplace=True)
df.rename(columns={' HIV/AIDS': 'HIV_AIDS_deaths'}, inplace=True)
df.rename(columns={'GDP': 'GDP_per_capita'}, inplace=True)
df.rename(columns={' thinness  1-19 years': 'Thinness_1_19_years'}, inplace=True)
df.rename(columns={' thinness 5-9 years': 'Thinness_5_9_years'}, inplace=True)
df.rename(columns={'Income composition of resources': 'Human_development_index'}, inplace=True)
df.rename(columns={'Schooling': 'Schooling_years'}, inplace=True)
```

```python
# Checking work

df.head()
```

```python
# Checking country names

df["Country"].value_counts()
```

```python
# Correcting country names

# List of country replacements
country_replacements = {
    'Micronesia (Federated States of)': 'Micronesia',
    'The former Yugoslav republic of Macedonia': 'Macedonia',
    'Venezuela (Bolivarian Republic of)': 'Venezuela',
    'Viet Nam': 'Vietnam',
    'Bolivia (Plurinational State of)': 'Bolivia',
    'Iran (Islamic Republic of)': 'Iran'
}

# Function to replace words in a given text
def replace_words(text):
    for old_word, new_word in country_replacements.items():
        text = text.replace(old_word, new_word)
    return text

# Apply the function to the specified column in the DataFrame
df['Country'] = df['Country'].apply(replace_words)

# Print the updated DataFrame
print(df)
```
